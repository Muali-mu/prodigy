```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'path_to_your_file.csv' with the actual file path)
data = pd.read_csv(r'C:\Users\SHIZA  SHABBIR\Downloads\tested (1).csv')

# Preview the dataset
print(data.head())
```

```python
# Extract relevant data: Age and Gender
age_data = data['Age'].dropna()  # Drop missing values for age
print(age_data)
```

```python
gender_data = data['Sex'].value_counts()  # Count occurrences of each gender
```

```python
# Extract relevantCreate a bar chart for gender distribution
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(gender_data.index, gender_data.values, color=['blue', 'pink'], edgecolor='black')
plt.title('Gender Distribution on Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python

# Create a histogram for age distribution
plt.subplot(1, 2, 2)
plt.hist(age_data, bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution on Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
